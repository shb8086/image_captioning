```python
import numpy as np
import pandas as pd
```

```python

IMAGES_CAPTIONS = '../dataset/caption.csv'

data_set = pd.read_csv(IMAGES_CAPTIONS)
data_set = data_set.to_numpy()
print(data_set[0][0])
captions = data_set[:,1]
# print(captions[100])
# print(len(captions))
```

```python
# Python program to find the k most frequent words
# from data set
K = 10

from collections import Counter
data_set_merged = ''.join(str(v) for v in captions)
# split() returns list of all the words in the string
split_it = data_set_merged.split()

# Pass the split_it list to instance of Counter class.
Counter = Counter(split_it)
  
# most_common() produces k frequently encountered
# input values and their respective counts.
most_occur = Counter.most_common(K)

# print(most_occur)  
# print(most_occur[0][0])

words = list()
counts = list()
for element in most_occur:
    words.append(element[0])
    counts.append(element[1])

print(words)
print(counts)
```

```python
import matplotlib.pyplot as plt
from bidi.algorithm import get_display
from arabic_reshaper import reshape

persian_labels = [get_display(reshape(word)) for word in words[:10]]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(persian_labels,counts[:10])
plt.xlabel('Word')
plt.ylabel('Repeats')
plt.savefig('most_common_words.png', bbox_inches="tight",pad_inches=0.3, transparent=True, dpi=200)
plt.show()
```

```python
def longest_char_size(some_list): 
    longest = 0    #You set the count to 0
    for i in range(len(some_list)): # Go through the whole list
        if len(some_list[i]) > longest: #Checking for the longest word(string)
            longest = len(some_list[i])
    return longest

def shortest_char_size(some_list): 
    shortest = 1000    #You set the count to 1000
    for i in range(len(some_list)): # Go through the whole list
        if len(some_list[i]) < shortest: #Checking for the longest word(string)
            shortest = len(some_list[i])
    return shortest

temp = [] #each caption word seperated
temp_len = [] #number of words in each caption 
for caption in captions:
    caption = caption.split(" ")
    temp.append(caption)
    temp_len.append(len(caption))

print("number of max chars:",longest_char_size(captions))
print("number of max words:",max(temp_len))
print(captions[temp_len.index(max(temp_len))])
# print("here" , data_set[temp_len.index(max(temp_len))][0])

shortest = shortest_char_size(captions)
print("number of min chars:",shortest)
print("number of min words:",min(temp_len))
print(captions[temp_len.index(min(temp_len))])

print('avg of words: {0:.2f}'.format(sum(temp_len)/len(temp_len)))
```
